```python
#creating empty list
brand=[]
model=[]
varient=[]
year=[]
kilometers=[]
fuel=[]
no_of_owners=[]
location=[]
price=[]
```

```python
for i in city_list:
    driver.get(i)
    #scrolling down to make sure we get more data in single frame
    #to increase the data counts
    start=0
    end=20
    for page in range(start,end):#for loop for scrapping 15 page
        #scrolling down to make sure we get 100 images in sinle frame
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        driver.execute_script("window.scrollTo(0,-1200)")
        time.sleep(5)
    #Extracting all href link
    href_url_1=[]
    url=driver.find_elements_by_xpath("//div[@class='card-detail-block__data']/a")
    for i in url:
        href_url_1.append(i.get_attribute("href"))
    for url in href_url_1:
        driver.get(url)
        #extracting price
        try:
            for i in driver.find_elements_by_xpath("/html/body/div[14]/section[2]/div[2]/div[2]/div/div[1]/div[1]/div/div[1]/p"):
                price.append(i.text.replace("Lakh","").replace("â‚¹","").replace(".","").replace(" ","")) #appending the price from price tag
        except:
            price.append("")
        #extracting brand tag
        try:
            for i in driver.find_elements_by_xpath("//ul[@class='special-skin-text']/li[4]/a/span"):
                brand.append(i.text)#appending the location from location tag
        except:
            brand.append("")
        #extracting model tag
        try:
            for i in driver.find_elements_by_xpath("//ul[@class='special-skin-text']/li[5]/a/span"):
                model.append(i.text)#appending the model from model tag
        except:
            model.append("")
        #extracting year tag
        try:
            for i in driver.find_elements_by_xpath("//*[@id='overview']/div/ul/li[2]/div[2]"):
                year_text=i.text.replace(" ","")
                year_regex=re.sub("[a-zA-Z]","",year_text)
                year.append(year_regex)#appending the year from year tag
        except:
            year.append("")
        #extracting kilometers tag
        try:
            for i in driver.find_elements_by_xpath("//*[@id='overview']/div/ul/li[3]/div[2]"):
                kilometers.append(i.text.replace("km","").replace(",","").replace(" ",""))#appending the kilometers from kilometer tag
        except:
            kilometers.append("")
        #extracting fuel tag
        try:
            for i in driver.find_elements_by_xpath("//*[@id='overview']/div/ul/li[4]/div[2]"):
                fuel.append(i.text)#appending the fuel type from fuel tag
        except:
            fuel.append("")
        #extracting no_of_owners tag
        try:
            for i in driver.find_elements_by_xpath("//*[@id='overview']/div/ul/li[9]/div[2]"):
                #appending the No of owners from no of owners tag
                no_of_owners.append(i.text.replace("First", "1st").replace("Second","2nd").replace("Third","3rd").replace("Fourth","4th").replace("or More","+"))
        except:
            no_of_owners.append("")
        #extracting location tag
        try:
            for i in driver.find_elements_by_xpath("//*[@id='available-text']"):
                loc_split=i.text.split()
                location.append(loc_split[-1])#appending the location from location tag
        except:
            location.append("")
price2=[]
price = [float(i) for i in price]
for i in price:
    price2.append(int(i*100000))
    

    
```
